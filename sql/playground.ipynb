```python

from vnstock3 import Vnstock

def create_sql_data(ticker: str, database_name: str = "financial_data.sqlite", source="TCBS"):
    retriever = Vnstock().stock(symbol=ticker, source=source)
    conn = sqlite3.connect(database_name)
    company = retriever.company
    # overview = company.overview()
    balance_sheet = process_finance_data(ticker, retriever.finance.balance_sheet(),conn,"balance_sheet")
    cash_flow = process_finance_data(ticker, retriever.finance.cash_flow(),conn,"cash_flow")
    income_statement = process_finance_data(ticker, retriever.finance.income_statement(),conn,"income_statement")
    ratio = process_ratio_data(ticker, retriever.finance.ratio(), conn)
    # overview = process_overview_data(ticker, overview)
    
    balance_sheet.to_sql("balance_sheet", conn, if_exists="append", index=False, dtype={'id': 'INTEGER PRIMARY KEY','ticker': 'TEXT'})
    cash_flow.to_sql("cash_flow", conn, if_exists="append", index=False, dtype={'id': 'INTEGER PRIMARY KEY','ticker': 'TEXT'})
    income_statement.to_sql("income_statement", conn, if_exists="append", index=False, dtype={'id': 'INTEGER PRIMARY KEY','ticker': 'TEXT'})
    ratio.to_sql("ratio", conn, if_exists="append", index=False, dtype={'id': 'INTEGER PRIMARY KEY','ticker': 'TEXT'})
    conn.close()

def get_max_id(conn, table_name):
    try:
        query = f"SELECT MAX(id) FROM {table_name};"
        max_id_result = pd.read_sql_query(query, conn)
        max_id = max_id_result.iloc[0, 0]  # Extract the max_id value
        return 0 if pd.isnull(max_id) else max_id
    except Exception as e:
        # print(f"Error getting max id for {table_name}: {e}")
        return 0

def process_finance_data(ticker, df, conn, table_name):
    max_id = get_max_id(conn, table_name)
    df.reset_index(inplace=True)
    df.rename(columns={'period': 'year'}, inplace=True)
    df['ticker'] = ticker
    df['id'] = df.index + 1 + max_id
    df['year'] = df['year'].astype(int)
    df['ticker'] = df['ticker'].astype('string')
    return df

def process_ratio_data(ticker, df, conn):
    max_id = get_max_id(conn, "ratio")
    df.reset_index(inplace=True)
    df.drop(columns=['period'], inplace=True)
    df['ticker'] = ticker
    df['id'] = df.index+1+max_id
    df['year'] = df['year'].astype(int)
    df['ticker'] = df['ticker'].astype('string')
    return df

create_sql_data("FPT")

conn = sqlite3.connect("financial_data.sqlite")

query_bs = pd.read_sql_query("SELECT * FROM balance_sheet limit 5", conn)
query_cf = pd.read_sql_query("SELECT * FROM cash_flow limit 5", conn)
query_is = pd.read_sql_query("SELECT * FROM income_statement limit 5", conn)

conn.close()
query_bs.head()
query_cf.head()
query_is.head()


```

```python
def get_table_schema(table_name: str, db_name: str = "financial_data.sqlite"):
    conn = sqlite3.connect(db_name)
    create_table_query = f"SELECT sql FROM sqlite_master WHERE type='table' AND name='{table_name}';"
    cursor = conn.execute(create_table_query)
    create_table_statement = cursor.fetchone()[0]
    
    conn.close()
    return create_table_statement


print(get_table_schema("balance_sheet"))
```
